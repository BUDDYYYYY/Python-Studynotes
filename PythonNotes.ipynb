## 3、标准差（Standard Deviation）
标准差：又常称为均方差，是一个数字，描述值的离散程度。<br>
低标准偏差表示大多数数字接近均值（平均值）。<br>
高标准偏差表示这些值分布在更宽的范围内。<br>



```python
import numpy as np
speed = [86,87,88,86,87,85,86]
# 计算标准差Sigma
x = np.std(speed)
# 计算方差Sigma Square
y = np.var(speed)
print("标准差为：",x)
print("方差为：",y)
```

## 8、线性回归（Linear Regression）
### Python提供一些方法查找数据点之间的关系并绘制线性回归线
### from W3School machine learning, linear regression

```python
# 导入所需模块
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 创建表示x和y轴的数值
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
# 执行一个方法，该方法返回线性回归的一些主要键值
slope,intercept,r,p,std_err = stats.linregress(x,y)
```

```python
# 创建一个使用 slope 和 intercept 值的函数返回新值。这个新值表示相应的 x 值将在 y 轴上放置的位置
def myfunc(x):
    return slope * x + intercept
```

```python
# 通过函数运行x数组的每个值，这将产生一个新的数组，其中的y轴具有新值
mymodel = list(map(myfunc,x))
```

```python
# 绘制原始散点图
plt.scatter(x,y)
# 绘制线性回归线
plt.plot(x,mymodel)
# 显示图
plt.show()
```

```python
# 回归的拟合度R-Squares
print(r)
```

```python
# 预测当x=10时的y值
y10=myfunc(10)
print(y10)
```

## 9、多项式回归（Polynomial Regression）
### 多项式回归使用变量x和y之间的关系来找绘制数据点线的最佳方法


```python
# 导入所需要模块
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# 创建x和y轴值的数组
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
# numpy里有一种方法能建立多项式模型
mymodel = numpy.poly1d(numpy.polyfit(x,y,3))
```

```python
# 指定行的显示方式，从位置1开始，到位置22结束
myline = numpy.linspace(1,22,100)
```

```python
# 绘制原始散点图
plt.scatter(x,y)
# 画出多项式回归线
plt.plot(myline,mymodel(myline))
# 显示图表
plt.show()
```

```python
# 多项式回归拟合度
print(r2_score(y,mymodel(x)))
```

```python
# 预测当x=17,时的y
y17 = mymodel(17)
print(y17)
```

## 13、决策树
### 决策树（Decision Tree）
展示如何制作“决策树”，决策树是一种流程图，可以根据以前的经验进行决策。<br>

这个例子中，一个人将借助决策树尝试决定他是否值得参加任何新的演出。



### 工作原理
首先，需要导入所需的模块，并使用pandas读取数据库

```python
import pandas as pd
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
df = pd.read_csv(r"C:\Users\11847\Desktop\Python\PythonNotes\shows.csv")
print(df)
```

制作决策树，需要所有数据都是数字。因此，需要将非数字列“Nationality”和“Go”转换为数值。<br>

Pandas库有一个map()方法，该方法接受字典，其中包含有关如何转换值的信息。<br>

{'UK':0, 'USA':1, 'N':2}<br>
表示：UK转换为0，USA转换为1，N转换为2


```python
d = {'UK':0, 'USA':1, 'N':2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES':1, 'NO':0}
df['Go'] = df['Go'].map(d)
print(df)
```

然后，我们必须将特征列和目标列分开；<br>
特征列：我们尝试从中预测的列；<br>
目标列：我们尝试预测的值的列。

```python
features = ['Age','Experience','Rank','Nationality']
X = df[features]
y = df['Go']
print(X)
print(y)
```

现在，可以创建实际决策树，使其适合我们的细节，然后在计算机上保存一个.png文件。

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)
data = tree.export_graphviz(dtree,out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img = pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```

### 结果解释
决策树使用您先前的决策来计算您是否愿意去看喜剧演员的几率<br>












                                

### 预测值
我们可以使用决策树来预测新值。<br>

例如：我是否应该去看一个由 40 岁的美国喜剧演员主演的节目，该喜剧演员有 10 年的经验，喜剧排名为 7？

```python
# 使用predict()方法预测新值
print(dtree.predict([[40,10,7,1]]))
print("[1] means 'Go'")
print("[0] means 'No'")
```

```python
# 如果喜剧等级为6呢？
print(dtree.predict([[40,10,6,1]]))
print("[1] means 'Go'")
print("[0] means 'No'")
```

# *如何从Python列表（List）中删除重复项



## Way1：列表转换为字典，再转换为列表

```python
#含重复项的List
mylist = ["a","b","a","c","c"] 
#使用列表项作为键 创建字典。这将自动删除任何重复项，因为字典不能有重复的键。
mylist = list(dict.fromkeys(mylist))  # 将字典转换回列表
print(mylist) #有一个没有任何重复的List，它与原始List有相同的顺序
```

## Way2：创建函数

创建一个函数，可以发送列表，然后返回它们的无重复项

```python
# 创建一个以List作为参数的函数
# 以此List项作为键创建字典
# 将字典转换为列表
def my_function(x):
    return list(dict.fromkeys(x))
mylist = my_function(["a","b","a","c","c"])
print(mylist)
```

# *如何反转Python中的字符串


## 创建切片

Python中没有内置函数可以反转字符串。<br>
最快的方法是使用切片，向后退步，-1


创建一个从字符串末尾开始的切片，然后向后移动。<br>

在这个特定的例子中，slice 语句 [::-1] 等同于 [11:0:-1]，这意味着从位置 11 开始（因为 "Hello "World" 有 11 个字符），结束于位置 0，移动步长 -1，负一意味着向后退一步。

```python
# 反转“Hello World”
txt = "Hello World"[::-1]
print(txt)
```

## 创建函数（带切片）

创建一个带上述切片方法的函数，只需要发送字符串，就能出反转字符串的结果

```python
def my_function(x):
    return x[::-1]
mytxt = "Hello World"
mytxt = my_function(mytxt)
print(mytxt)
```
